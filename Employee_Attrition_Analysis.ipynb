```python
sns.countplot(x='Attrition', data=df)
plt.title('Employee Attrition Count')
plt.show()
```

```python
#percentage of attrition
attrition_rate = df['Attrition'].value_counts(normalize=True) * 100
print(attrition_rate)
```

```python
plt.figure(figsize=(8,5))
sns.boxplot(x='Attrition', y='Age', data=df)
plt.title('Age vs Attrition')
plt.show()
```

```python
plt.figure(figsize=(8,5))
sns.countplot(x='Department', hue='Attrition', data=df)
plt.title('Department vs Attrition')
plt.xticks(rotation=45)
plt.show()
```

```python
plt.figure(figsize=(10,5))
sns.countplot(x='JobRole', hue='Attrition', data=df)
plt.title('Job Role vs Attrition')
plt.xticks(rotation=45)
plt.show()
```

```python
plt.figure(figsize=(8,5))
sns.boxplot(x='Attrition', y='MonthlyIncome', data=df)
plt.title('Monthly Income vs Attrition')
plt.show()
```

```python
plt.figure(figsize=(8,5))
sns.boxplot(x='Attrition', y='YearsAtCompany', data=df)
plt.title('Years at Company vs Attrition')
plt.show()
```

A correlation heatmap is a graphical representation of a correlation matrix. It uses colors to show the strength and direction of the linear relationship between pairs of variables.

```python
plt.figure(figsize=(12,8))
# Select only numeric columns for correlation calculation
numeric_df = df.select_dtypes(include=np.number)
sns.heatmap(numeric_df.corr(), cmap='coolwarm', annot=False)
plt.title('Correlation Heatmap')
plt.show()
```

# Key Insights
1.   Most employees who left are younger (25–35 years).
2.   Sales Department shows higher attrition.
3.   Employees with lower income tend to leave more.
4.   Those who work overtime have a higher attrition rate.
5.   Attrition is higher among employees with 1–3 years at the  company.













# Conclusion
The analysis successfully identified key factors influencing employee attrition using the IBM HR dataset.
Data revealed that younger, lower-paid employees in sales and overtime-heavy roles are most likely to leave the company.
These insights can help HR teams focus on employee engagement and retention strategies to reduce turnover.
